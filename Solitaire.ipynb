```python
from random import randint

def swap_two_random_cards(deck):
    first_card_index = randint(0, 39)
    second_card_index = randint(0, 39)
    deck[first_card_index], deck[second_card_index] = deck[second_card_index], deck[first_card_index]
```

```python
deck
```

```python
swap_two_random_cards(deck)
```

```python
for i in range(100):
    swap_two_random_cards(deck)
```

```python
def shuffle_deck(deck):
    for i in range(100):
        swap_two_random_cards(deck)
```

```python
def p_win(cl, n_exp):
    return sum([match(cl) for _ in range(n_exp)])/n_exp
```
