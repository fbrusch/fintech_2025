```python
from random import randint
def swap_two_random_elements_of_list(l):
    x = randint(0, len(l)-1)
    y = randint(0, len(l)-1)
    l[x], l[y] = l[y], l[x]
    
    
```

```python
def play_a_complete_round():
    shuffle(doors)
    chosen_door = int(input("please choose a door (0, 1, 2)"))
    # I have to show the user a door that:
    # 1. is different from the one he chose
    # 2. doesn't hide a prize
    if doors[0] == 0 and chosen_door != 0:
        print("Behind door 0 there is no prize")
    elif doors[1] == 0 and chosen_door != 1:
        print("Behind door 1 there is no prize")
    elif doors[2] == 0 and chosen_door != 2:
        print("Behind door 2 there is no prize")
        
    chosen_door = int(input("you can chose your door anew (0, 1, 2)"))
    if doors[chosen_door] == 1:
        print("congrats, you won!")
    else:
        print("sorry, you lost..:")
    print("behind doors there was: ", doors)
```

```python
def remaining_door(door_chosen, door_opened):
    return 3 - (door_chosen + door_opened)

def play_a_complete_round_changing():
    global rounds_played, rounds_won
    shuffle(doors)
    chosen_door = randint(0, 2)
    print("you chose door", chosen_door)
    # I have to show the user a door that:
    # 1. is different from the one he chose
    # 2. doesn't hide a prize
    if doors[0] == 0 and chosen_door != 0:
        print("Behind door 0 there is no prize")
        opened_door = 0
    elif doors[1] == 0 and chosen_door != 1:
        print("Behind door 1 there is no prize")
        opened_door = 1
    elif doors[2] == 0 and chosen_door != 2:
        print("Behind door 2 there is no prize")
        opened_door = 2

    # we should now chose the "remaining" door: not the one we chose, 
    # not the one that was opened
    chosen_door = remaining_door(chosen_door, opened_door)

    # chose_remainig(2,1) -> 0 
    
    print("2nd chance, you chose door", chosen_door)
    
    if doors[chosen_door] == 1:
        print("congrats, you won!")
        rounds_won += 1
    else:
        print("sorry, you lost..:")
    print("behind doors there was: ", doors)
    rounds_played += 1
```

say we want to play with 4 doors!

```python
from random import choice

def remaining_door(door_chosen, door_opened, n_of_doors):
    possible_doors = [x for x in range(n_of_doors) if x != door_chosen and x != door_opened]
    return choice(possible_doors)


def play_a_complete_round_changing_4_doors():
    global rounds_played, rounds_won
    n_of_doors = 4
    doors = [0] * (n_of_doors - 1) + [1]
    shuffle(doors)
    chosen_door = randint(0, n_of_doors - 1)
    print("you chose door", chosen_door)
    
    possible_doors_to_open = [x for x in range(n_of_doors) if x != chosen_door and doors[x] == 0]
    opened_door = choice(possible_doors_to_open)
    print("Behind door", opened_door, "there is no prize")
    
    # we should now chose the "remaining" door: not the one we chose, 
    # not the one that was opened
    chosen_door = remaining_door(chosen_door, opened_door, n_of_doors)

    # chose_remainig(2,1) -> 0 
    
    print("2nd chance, you chose door", chosen_door)
    
    if doors[chosen_door] == 1:
        print("congrats, you won!")
        rounds_won += 1
    else:
        print("sorry, you lost..:")
    print("behind doors there was: ", doors)
    rounds_played += 1
```
