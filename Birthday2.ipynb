```python
def count_elements(l, n):
    counter = 0
    for i in l:
        if i == n:
            counter = counter + 1

    return counter

            
```

```python
count_elements([1,2,3,3,3,4], 3)
```

```python
count_elements([1,2,2,3,3,3,2,1,23], 10)
```

```python
class_of_people = [generate_proper_birthday() for i in range(30)]
```

```python
print(class_of_people)
```

```python
count_elements(class_of_people, 58)
```

```python
birthday_freqs = [count_elements(class_of_people, x) for x in range(366)]
```

```python
len([x for x in birthday_freqs if x > 1]) > 0
```

```python
def check_if_there_are_two_same_elements(l):
    birthday_freqs = [count_elements(l, x) for x in range(366)]
    return len([x for x in birthday_freqs if x > 1]) > 0
```

```python
check_if_there_are_two_same_elements(class_of_people)
```

```python
def create_class_of_people():
    return [generate_proper_birthday() for i in range(30)]
```

```python
sum([check_if_there_are_two_same_elements(create_class_of_people()) for i in range(100)])
```
