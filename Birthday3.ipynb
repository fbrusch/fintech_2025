```python
def generate_birthday():
    """this function returns a number between 1 and 365, 
       365 representing the 29-th of february"""
    days = list(range(365)) * 4 + [365]
    return choice(days)
```

```python
def generate_class_of_people(number_of_persons):
    return [generate_birthday() for x in range(number_of_persons)]
```

```python
generate_class_of_people(30)
```

```python
def count_elements(l, value):
    return len([x for x in l if x == value])
```

```python
count_elements([1,2,3,4], 4)
```

```python
class_of_people = generate_class_of_people(30)
```

```python
count_elements(class_of_people, 3)
```

```python
birthday_frequencies_for_day = [count_elements(class_of_people, x) for x in range(366)]
```

```python
len([x for x in birthday_frequencies_for_day if x > 1]) > 0
```

```python
def check_multiple_birthdays(people):
    birthday_frequencies_for_day = [count_elements(people, x) for x in range(366)]
    result = len([x for x in birthday_frequencies_for_day if x > 1]) > 0
    return result
```

```python
c = generate_class_of_people(30)
```

```python
check_multiple_birthdays(c)
```

```python
def experiment():
    return check_multiple_birthdays(generate_class_of_people(30))
```

```python
def estimate_same_birthday_probability(n_of_experiments=100):
    return sum([experiment() for x in range(n_of_experiments)])/n_of_experiments
```

```python
estimate_same_birthday_probability(10000)
```
